1. Write python code to use the trapezoidal rule to intergrate exp(-x^2)from x= 0 to  x= 1

```python
import numpy as np
#define the function and interval
a, b = 0, 1
n = 10
x = np.linspace(a, b, n + 1)
def f(x):
    return np.exp(-x**2)
h = (b - a) / n
integral = h / 2 * (f(a) + 2 * sum(f(x[1:n])) + f(b))
print(f"Trapezoidal Rule Approximation: {integral:.6f}")
```

2. write python code to use Euler's method to solve dy/dx = exp(-x^2) with the initial condtion y(0)=1

```python
import numpy as np
h = 0.1  
num_steps = 10
x0 = 0
y0 = 1
xs = [x0]
ys = [y0]
def euler_method(f, x0, y0, h, num_steps):
    x, y = x0, y0
    for _ in range(num_steps):
        y += h * f(x)
        x += h
        xs.append(x)
        ys.append(y)
    return xs, ys
for xi, yi in zip(xs, ys):
    print(f"x: {xi:.2f}, y: {yi:.6f}")
print(xs)
print(ys)
```
